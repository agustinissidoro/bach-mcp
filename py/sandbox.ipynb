```python
from tcp import TCPSend, TCPServer
import time

# Setup: Start Python server and sender
print("=== SETUP ===\n")

server = TCPServer()
server.set_default_handler(lambda msg: print(f"FROM MAX: {msg}"))
server.start()

sender = TCPSend()

time.sleep(0.5)
print("\n=== READY ===\n")
```

```python
print("=== TEST: Send to Node.js/Max ===\n")
sender.send("Hello from Python")
time.sleep(0.5)
```

```python
print("=== TEST: Send JSON to Node.js/Max ===\n")
sender.send_dict({"message": "Test JSON", "value": 42})
time.sleep(0.5)
```
